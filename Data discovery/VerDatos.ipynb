```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/cior666/Machine-Learning-III-entrega/refs/heads/main/Data_preprocessing/car_pricing.csv")
```

```python
df
```

```python
df.info()
```

```python
#voy a revisar el porcentaje de nulos por columna
df.isna().sum()/df.shape[0]*100
```

```python
#como notrepaireddamage tiene muchos nulos y lo considero importante para mi futuro analisis eda voy a rellenar
df.notRepairedDamage.value_counts()
```

```python
#genero los random con los que rellenare
random_nRD=np.random.choice(["nein","ja"],size=df.notRepairedDamage.isnull().sum())
```

```python
#relleno
df.loc[df["notRepairedDamage"].isnull(),"notRepairedDamage"]=random_nRD
```

```python
#cambio los ja y nein 
df['notRepairedDamage'] = df['notRepairedDamage'].replace({'ja': 'si', 'nein': 'no'})
```

```python
#Para las columnas con nulos voy a rellenar los valores usando un random generate.
df.vehicleType.value_counts()
```

```python
#Procedo a rellenar los null con las 3 alternativas más repetidas
random_vtype=np.random.choice(["limousine","kleinwagen","kombi"],size=df.vehicleType.isnull().sum())
```

```python
#relleno
df.loc[df["vehicleType"].isnull(),"vehicleType"]=random_vtype
```

```python
#Hago lo mismo para model
df.model.value_counts()
```

```python
#mismo analisis y procedimiento para model
random_model=np.random.choice(["golf","andere","3er","polo"],size=df.model.isnull().sum())
```

```python
#relleno
df.loc[df["model"].isnull(),"model"]=random_model
```

```python
#Ahora lo hacemos para gearbox
df.gearbox.value_counts()
```

```python
#procedo igual que los casos anteriores
random_gearbox=np.random.choice(["manuell","automatik"],size=df.gearbox.isnull().sum())
```

```python
#relleno
df.loc[df["gearbox"].isnull(),"gearbox"]=random_gearbox
```

```python
#ahora para fueltype
df.fuelType.value_counts()
```

```python
#genero con los 3 mas repetidos
random_fueltype=np.random.choice(["benzin","diesel","lpg"],size=df.fuelType.isnull().sum())
```

```python
#relleno
df.loc[df["fuelType"].isnull(),"fuelType"]=random_fueltype
```

```python
#verificamos los null de nuevo
df.isna().sum()/df.shape[0]*100
```

```python
#como buena practica verifico que no haya duplicados
df.duplicated().sum()
```
